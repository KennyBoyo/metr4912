```python
import numpy as np
```

```python
def generate_data(n_points, seed=0):
    return np.random.uniform(-np.pi, np.pi, (n_points, 3))

data = generate_data(10)
bins = np.linspace(-np.pi, np.pi, 100)

print(bins)

x_indices = np.digitize(data, bins)
print(x_indices)
```
