```python
import numpy as np
import open3d as o3d
import plotly.graph_objects as go
import plotly
```

```python
def generate_data(n_points, seed=0):
    np.random.seed(0)
    return np.random.uniform(-1, 1, (n_points, 3))

data = generate_data(10000)
x = data[:, 0]
y = data[:, 1]
z = data[:, 2]
```

```python
fig = go.Figure(data=[go.Mesh3d(x=x, y=y, z=z, alphahull=4, color='lightpink', opacity=0.50)])
fig.show()
```

```python
coord_pipe = 'coord_pipe'

with open(coord_pipe, 'rb') as f:
    coord_array = np.loadtxt(f)

fig = go.Figure(data=[go.Mesh3d(x=coord_array[:, 0], y=coord_array[:, 1], z=coord_array[:, 2], alphahull=20, color='lightpink', opacity=0.50)])
fig.show()
print(coord_array)
```
